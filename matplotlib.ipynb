```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#this will allow to see the plots in the notebook
```

```python
import numpy as np
```

```python
X = np.linspace(0,5,11)
Y = X**2
```

```python
X
```

```python
Y
```

```python
# functional method
plt.plot(X,Y, marker="X")
```

```python
Z=np.sin(X)
```

```python
plt.plot(X,Z,'r-')
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(X,Z,'r-')
```

```python
plt.subplot(1,2,1)
plt.plot(X,Y,'r')

plt.subplot(1,2,2)
plt.plot(Y,X,'b')
```

```python
# 00
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.plot(X,Y, marker="X")
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
# 00
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(X,Y, marker="X")
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.05,0.4,0.4,0.3])

axes1.plot(X,Y, marker="X")
axes1.set_xlabel("X label")
axes1.set_ylabel("Y label")
axes1.set_title("Title X Y")

axes2.plot(X,Z, marker="X")
axes2.set_xlabel("X label")
axes2.set_ylabel("Z label")
axes2.set_title("Title X Z")
```

```python
fig,axes3 = plt.subplots(1,2)
i=1
for current in axes3:
    current.plot(X,Y)
    title = "title"+str(i)
    current.set_title(title)
    i+=1
```

```python
fig,ax = plt.subplots(1,2,figsize=(8,2))

ax[0].plot(X,Y)
ax[0].set_xlabel("x")
ax[0].set_ylabel("x^2")
ax[0].set_title("y = x^2")


ax[1].plot(X,Z)
ax[1].set_xlabel("x")
ax[1].set_ylabel("sin x")
ax[1].set_title("y = sin x")
```

```python
fig.savefig("my_picture.png",dpi=200)
```

```python
# 00
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(X,Y, marker="X",label='X^2')
axes.plot(X,Z, marker="X",label = 'sinX')
axes.legend(loc=(0.1,0.8))
```

```python
fig = plt.figure()
axes4 = fig.add_axes([0,0,1,1])
axes4.plot(X,Y,color = 'green', lw=1, marker='o', markersize=20, markerfacecolor='yellow' )
axes4.plot(X,Z,color = 'blue',linestyle='-', lw=1,marker='o', markersize=20)
axes4.set_xlim([0,4])
axes4.set_ylim([0,4])
```
